```python
from matplotlib import pyplot as plt
from matplotlib import style

x = [0, 1, 2, 3, 4, 5]
y = [24.5, 23.18, 25.9, 30.5, 22.3, 24.6]

fig, ax = plt.subplots()

ax.bar(x, y, align='center')

ax.set_title('Bahasa pemrograman favorit')
ax.set_ylabel('Pengguna')
ax.set_xlabel('Bahasa')

ax.set_xticks(x)
ax.set_xticklabels(("Phyton", "java", "javascript", "C#", "php", "c++"))

plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Teknisi', 'Guru', 'Pegawai Kantor', 'Pedagang']
quantity = [3, 10, 12, 11]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']

plt.title('Daftar Pekerjaan lulusan Darmajaya')
plt.pie(quantity, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)

plt.axis('equal')
plt.show()
```
